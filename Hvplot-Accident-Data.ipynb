```python
import hvplot.pandas
import pandas as pd

accident_csv = "./accident-long and lat clean.csv"
accident_df = pd.read_csv(accident_csv)
```

```python
map_plot = accident_df.hvplot.points(
    "LONGITUD",
    "LATITUDE",
    geo=True,
    size="PERSONS",
    scale=3,
    color="county_name",
    alpha=0.5,
    tiles =True,
    frame_width=1000,
    frame_height=1000
)
map_plot
```

```python
map_plot = accident_df.hvplot.points(
    "LONGITUD",
    "LATITUDE",
    geo=True,
    size=2,
    scale=2,
    color="red",
    alpha=0.5,
    tiles =True,
    frame_width=1000,
    frame_height=1000
)
map_plot
```
