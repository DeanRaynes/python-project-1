```python
#Read data from CSV files on Accidents
accident_csv = "./accident.csv"
person_csv = "./person.csv"
vehicle_csv = "./vehicle.csv"

accident_df = pd.read_csv(accident_csv)
person_df = pd.read_csv(person_csv)
vehicle_df = pd.read_csv(veh,icle_csv)
```

```python
columnnames = ['accident_id','CITY','city_name','COUNTY','county_name','LATITUDE','LONGITUD'] 
location_df = accident_df[columnnames].copy()

location_df

accident_counts_per_county_df = pd.DataFrame(location_df['county_name'].value_counts())
accident_counts_per_county_df

per_county_spending_df["Accidents"] = [59,75,127,62,18,109,40,87,11,97,25,221,13,153,757]
per_county_spending_df["Accidents Per Capita"] = ""
per_county_spending_df["Spending in Billions"] = ""
per_county_spending_df["Per Capita in 10K"] = ""
per_county_spending_df["Accidents Per Capita in 10kths"] = ""
for i in range(15):
    per_county_spending_df.loc[i,"Accidents Per Capita"] = per_county_spending_df.loc[i,"Accidents"]/per_county_spending_df.loc[i,"Population"]
    per_county_spending_df.loc[i,"Spending in Billions"] = per_county_spending_df.loc[i,"Total Spending"]/1000000000
    per_county_spending_df.loc[i,"Per Capita in 10K"] = per_county_spending_df.loc[i,"Per Capita"]/10000
    per_county_spending_df.loc[i,"Accidents Per Capita in 10kths"] = per_county_spending_df.loc[i,"Accidents Per Capita"]*10000
    
per_county_spending_df
```

```python
sorted_spending_with_accident_df = per_county_spending_df.sort_values("Population", ascending = False)
sorted_spending_with_accident_df
```

```python
#Plot Total Spending with Accidents by County

fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.35
index = range(len(sorted_spending_with_accident_df))

bar1 = ax.bar(index, sorted_spending_with_accident_df['Spending in Billions'], bar_width, label='Spending in Billions')
bar2 = ax.bar([i + bar_width for i in index], sorted_spending_with_accident_df['Accidents'], bar_width, label='Accidents')

ax.set_xlabel('County')
ax.set_ylabel('Values')
ax.set_title('Total Spending and Accidents by County')
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(sorted_spending_with_accident_df['County'], rotation=45, ha='right')
ax.legend()

plt.tight_layout()
plt.savefig('Total__Spending_and_Accidents.png', format='png', dpi=300)
plt.show()

```
